## Types of Queue

###  Circular Queue

```python
# Types Of Queue
class CircularQueue():
    
    def __init__(self, k):
        self.k = k                        # k = Size
        self.queue = [None] * k           # self.queue = [None for i in range(k)]
        self.front = self.rear = -1
        
    # Insertion of elements in a Circular Queue
    def enqueue(self, data):
        
        # Condition if Queue is Full
        if ((self.rear + 1) % self.k == self.front):
            print('The Queue is Full...!')
        
        # Condition For Empty Queue
        elif (self.front == -1):
            self.front = 0
            self.rear = 0
            self.queue[self.rear] = data
            
        else :
            # Next position rare
            self.rear = (self.rear + 1) % self.k
            self.queue[self.rear] = data
        
        
    # Deletion of elements from a Circualr Queue
    def dequeue(self):
        
        # Condition for empty Queue
        if (self.front == -1):
            print('The Circular Queue is Empty...!')
            
        # Condition for only one element
        elif (self.front == self.rear):
            temp = self.queue[self.front]
            self.front = -1
            self.rear = -1
            return  temp
        
        else :
            temp = self.queue[self.front]
            self.front = (self.front + 1) % self.k
            return temp
    
    def display(self):
        if (self.front == -1):
            print('The Circular queue doesn\'t have an element')
        
        elif (self.rear >= self.front):                         # [][][][][]
            for i in range(self.front, self.rear+1):                   
                print(self.queue[i], end=' ')
            print()                                      # print() to print a blank line
        else:
            for i in range(self.front, self.k):
                print(self.queue[i], end=' ')
            for i in range(0, self.rear+1):
                print(self.queue[i], end = ' ')
        
obj = CircularQueue(6)
obj.enqueue(1)
obj.enqueue(2)
obj.enqueue(3)
obj.enqueue(4)
obj.enqueue(5)
obj.enqueue(9)
print('Initial Queue')
obj.display()

obj.dequeue()
print('\nAfter Deleting an element from the Queue')
obj.display()
```

### Priority Queue

```python
# Priority Queue

# Using Heap Data Structure
# MAx value is Top Priority


class PriorityQueue:
    
    def __init__(self):
        self.__queue = [0]
        self.__size = 0
        
        
    def __swap_up(self, i:int)-> None:
        
        while (i)//2 > 0:
            if self.__queue[i] > self.__queue[i//2]:
                self.swap(self.__queue, i, i//2)
            i //= 2
            
                
    def insert(self, value:int) -> None:
        self.__queue.append(value)
        self.__size += 1
        self.__swap_up(self.__size)
        
        
    def __swap_down(self, i:int) -> None:
        
        while self.__size >= 2 * i :
            if 2*i > self.__size:
                bigger_child = 2 * i 
            
            else :
                if self.__queue[2*i] > self.__queue[2*i+1]:
                    bigger_child = 2*i
                else:
                    bigger_child = 2*i+1

            if self.__queue[i] < self.__queue[bigger_child]:
                self.swap(self.__queue, i,bigger_child)
            i = bigger_child
                
            
                        
    def pop(self, value:int) -> None:
        max_value = self.__queue[1]
        i = self.__queue.index(value)
        self.swap(self.__queue, i,self.__size)
        self.__size -= 1
        self.__queue.pop()
        self.__swap_down(i)
        return max_value
    
    @property
    def get_list(self):
        return self.__queue[1:]
    
    def __len__(self):
        return self.__size
    
    def swap(self, L, i, j ):
        L[i], L[j] = L[j], L[i]
    
    
if __name__ == "__main__":
    import doctest

    doctest.testmod()
    # create an instance of BinaryHeap
    m = PriorityQueue()
    m.insert(63)
    m.insert(78)
    m.insert(12)
    m.insert(25)
    m.insert(1)
    m.insert(17)
    m.insert(11)
    m.insert(212)
    m.pop(78)
    m.insert(312)
    m.pop(17)
#     m.pop(312)
#     m.pop(212)
    m.insert(43)
    print(m.get_list)   
        
```

### Doubly-Ended Queue

```python
# Deque (Double ended Queue) Using Circular Array

class Deque:
    
    def __init__(self):
        self.items = []
    
    def isEmpty(self):
        if (self.items == []) :
            print('Given Deque is Empty...!')
        
        else:
            print('Nah..!')
    
    def addRear(self, item):
        self.items.append(item)
        
    def addFront(self, item):
        self.items.insert(0, item)
        
    def removeFront(self):
        self.items.pop()
        
    def removeRare(self):
        self.items.pop(0)
        
    def size(self):
        return len(self.items)
    
    
d = Deque()
print(d.isEmpty())
d.addRear(8)
d.addRear(5)
d.addFront(7)
d.addFront(10)
print(d.size())
print(d.isEmpty())
d.addRear(11)
print(d.removeRare())
print(d.removeFront())
d.addFront(55)
d.addRear(45)
print(d.items)
```

## 3. Linked Lists

```python
# Linked Lists

# 1. Singly LinkedList

class Node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList():
    
    def __init__(self):
        self.head = None
        
    def InsertAtBegining(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        
    def InsertAnywhere(self, position, data):               # Position = index
        new_node = Node(data)
        
        if position is None:
            print('The given Previous node must be in LinkedList')
            return
        
        temp = self.head
        for i in range(position- 1):
            temp = temp.next
            
        new_node.next = temp.next    
        temp.next  = new_node
        
        
        
    def InsertAtEnd(self, data):
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
            return
        
        last = self.head
        
        while (last.next):
            last = last.next
            
        last.next = new_node
        
    
    def length(self):
        current = self.head
        total = 0
        
        while current.next:
            total += 1
            current = current.next
            
        return total
    
    def DeletNode(self, position):
        
        if self.head == None:
            print("The LinkedList is empty...!")
        
        temp_node = self.head
        
        if position == 0:
            
            temp_node.next = self.head
            temp_node = None
            return
        
        for i in range(position):                  # Find the key to be deleted
            temp_node = temp_node.next
            
            if temp_node is None:
                break
                
        if temp_node is None:                          # If the key is not present
            return
        if temp_node.next is None:
            return
        
        next = temp_node.next.next
        temp_node.next = None
        temp_node.next = next
        
        
    def Print_List(self):
        
        temp_node = self.head
        
        while(temp_node):
            print(str(temp_node.data) + " ", end = "")
            temp_node = temp_node.next
            
if __name__ == '__main__':
        
    llist = LinkedList()
        
    llist.InsertAtEnd(12)
    llist.InsertAtBegining(2)
    llist.InsertAtBegining(3)
    llist.InsertAtEnd(4)
    llist.InsertAtEnd(8)
    llist.InsertAtBegining(10)
    llist.InsertAnywhere(3, 21)
            
    print('Linked List:')
    llist.Print_List()
    
    print('\n After Deleting A Node')
    llist.DeletNode(3)
    llist.Print_List()
    
```

### Doubly Linked List

```python
# 2. Doubly Linked List

class Node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class DoublyLL:
    
    def __init__(self):
        self.head = None
        
    def InsertAtBeginning(self, data):
        
        new_node = Node(data)
        new_node.next = self.head
        new_node.prev = None
        self.head = new_node
    
    def InsertAtEnd(self, data):
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while (temp.next):
            temp = temp.next
            
        temp.next = new_node
        new_node.prev = temp
        new_node.next = None
        
    def InsertAnywhere(self, index, data):
        new_node = Node(data)
        
        if index is None:
            raise ValueError('Enter any Valid position')
            return
        
        temp = self.head
        for i in range(index-1):
            temp = temp.next
            
        new_node.next  = temp.next
        new_node.prev = temp
        temp.next = new_node
        
    def DeletNode(self, position):
        
        if self.head is None:
            return
        
        temp = self.head
        
        if position == 0:
            self.head = temp.next
            temp = None
            return
        
        for i in range(position-1):
            temp = temp.next
        if temp is None:
            return
        
        if temp.next is None:
            return
        
        next = temp.next.next
        temp.next = None
        temp.next = next
        next.prev = temp
        
    def Print(self):
        temp_node = self.head
        while(temp_node):
            print(str(temp_node.data) + " ", end = "")
            temp_node = temp_node.next
        print()
                
    def Size(self):
        temp_node = self.head
        count = 0
        while(temp_node):
            temp_node = temp_node.next
            count += 1
        return count
            
l = DoublyLL()
l.InsertAtBeginning(54)
l.InsertAtBeginning(78)
l.InsertAtBeginning(51)
l.InsertAtBeginning(24)
l.InsertAtEnd(12)
l.InsertAtEnd(124)
l.InsertAnywhere(5, 58)
l.DeletNode(5)
l.Print()
```

```python
#CircularLinkedList

class Node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
        
class CircularLL:
    
    def __init__(self):
        self.head = None
        self.tail = None
        
    
    def InsertAtBeginning(self, data):
        new_node = Node(data)
        
        new_node.next = self.head
        temp = self.head
        if (temp == None):
            self.head = new_node
            self.tail = new_node
            self.head.next = new_node
            return
        
        if self.tail != None:
            self.tail.next = new_node
            new_node.next = self.head
            self.head = new_node
            return
        
        
    def InsertAtEnd(self, data):
        new_node = Node(data)
        
        if (self.head == None) and (self.tail == None):
            self.head = new_node
            self.head.next  = new_node
            self.tail = new_node
            return
        
        if self.tail != None:
            self.tail.next = new_node
            new_node.next = self.head
            self.tail = new_node
            
            
    def Delete(self, index):
        
        temp = self.head
        
        if temp is None:
            print('The Linked List is Empty....!')
        
        if index == 0:
            self.head = temp.next
            self.tail.next = temp.next
            temp = None
            return
                
        for _ in range(index - 1):
            temp = temp.next
        
            if temp.next is self.head:
                break
                
        next = temp.next.next
        temp.next = None
        temp.next = next   
        return
        
    def Print(self):
        temp = self.head
            
        while temp.next:
            print(str(temp.data), end = " ")
            temp = temp.next
                
            if temp == self.head:
                break
                    
        print()
                
            
            
        
#     def InsertAnywhere(self, data):

cll = CircularLL()
cll.InsertAtBeginning(87)
cll.InsertAtEnd(88)
cll.InsertAtBeginning(86)
cll.InsertAtEnd(12)
cll.InsertAtEnd(66)
cll.InsertAtEnd(45)
cll.InsertAtBeginning(32)
cll.InsertAtBeginning(22)
cll.Delete(4)
cll.Delete(0)
cll.Print()
```

### Open Addressing

```python
# Handling Collisions

# 2. Open Addressing

#### A. Linear Probing

class HashTable:
    
    def __init__(self, size):
        self.size = size
        self.keys = [None] * self.size
        self.values = [None] * self.size
        
    def Hash_Function(self, key):
        return key % self.size
    
    def Insert(self, key, value):
        
        index = self.Hash_Function(key)
        
        while self.keys[index] is not None:
            
            if self.keys[index] == key:
                self.values[index] = value
                return
            
            index = (index + 1) % self.size       #### B. For Quadratic Probing = (index**2 + 1) % self.size
            
        self.keys[index] = key    
        self.values[index] = value
        
        
    def Delete(self, key, value):
        index = self.Hash_Function(key)
        self.keys[index] = None
        self.values[index] = None

        
              
    def Valueof(self, key):
        
        index = self.Hash_Function(key)
        
        while self.keys[index] is not None:
            if self.keys[index] == self.keys:
                return self.values[index]
            
            index = (index + 1) % self.size
            
        return None
    
    def Print(self):
        
        for i in range(self.size):
            print(self.keys[i], end = ' --> ')
            print(self.values[i])
    
h = HashTable(10)
h.Insert(43, 'a')
h.Insert(14, 'b')
h.Insert(31, 'c')
h.Insert(87, 'd')
h.Insert(21, 'e')
h.Insert(29, 'f')
h.Insert(45, 'g')
h.Insert(83, 'h')
h.Insert(42, 'i')
h.Insert(32, 'j')
# h.Delete(43, 'Abhi')
h.Print()
```

### Max Heap

```python
#Heap Data Structure

# 1.Max Heap

class MAXHEAP:
    
    def __init__(self):
        self.__heap = [0]
        self.__size = 0
        
        
    def __swap_up(self, i:int)-> None:
        
        while (i)//2 > 0:
            if self.__heap[i] > self.__heap[i//2]:
                self.swap(self.__heap, i, i//2)
            i //= 2
            
                
    def insert(self, value:int) -> None:
        self.__heap.append(value)
        self.__size += 1
        self.__swap_up(self.__size)
        
        
    def __swap_down(self, i:int) -> None:
        
        while self.__size >= 2 * i :
            if 2*i == self.__size:
                bigger_child = 2 * i 
            
            else :
                if self.__heap[2*i] > self.__heap[2*i+1]:
                    bigger_child = 2*i
                else:
                    bigger_child = 2*i+1

            if self.__heap[i] < self.__heap[bigger_child]:
                self.swap(self.__heap, i,bigger_child)
            i = bigger_child
                
            
                        
    def Delete(self, value:int) -> None:
        i = self.__heap.index(value)
        self.swap(self.__heap, i,self.__size)
        self.__size -= 1
        self.__heap.pop()
        self.__swap_down(i)
        
    def Max(self):
        return self.__heap[1]
    
    @property
    def get_list(self):
        return (self.__heap[1:])
    
    def __len__(self):
        return self.__size
    
    def swap(self, L, i, j ):
        L[i], L[j] = L[j], L[i]
    
    
if __name__ == "__main__":
    import doctest

    doctest.testmod()
    # create an instance of BinaryHeap
    m = MAXHEAP()
    m.insert(63)
    m.insert(78)
    m.insert(12)
    m.insert(25)
    m.insert(1)
    m.insert(17)
    m.insert(11)
    m.insert(212)
    m.Delete(78)
    m.insert(312)
    m.Delete(17)
    m.Delete(312)
    m.insert(43)
    m.insert(64)
    print(m.get_list)
    print(len(m))
    print(m.Max())
```

```python
# Min Heap

class Node:
    
    def __init__(self, name, value):
        self.name = name
        self.value = value
        
class MINHEAP:
    
    def __init__(self):
        self.heap = [0]
        self.size = 0
        
    def Max(self):
        return self.heap[1]
    
    @property
    def get_list(self):
        return (self.heap[1:])
    
    def __len__(self):
        return self.size
    
    def swap(self, L, i, j ):
        L[i], L[j] = L[j], L[i]
        
        
    def Swap_up(self, i:int) -> None:
        
        while i//2 > 0:
            
            if self.heap[i] < self.heap[i//2]:     # i = Child, i//2 = Parent
                self.swap(self.heap, i, i//2)
            i //= 2
            
            
    def Swap_down(self, i:int):
        
        while self.size >= 2 * i:
            if 2 * i == self.size:
                smaller_child = 2*i
                
            else:
                if self.heap[2*i] < self.heap[2 * i + 1]:
                    smaller_child = 2*i
                    
                else:
                    smaller_child = 2*i+1
                    
            if self.heap[i] > self.heap[smaller_child]:
                self.swap(self.heap, i, smaller_child)
            
            i = smaller_child
                    
    
      
        
    def insert(self, value:int) -> None:
        self.heap.append(value)
        self.size += 1
        self.Swap_up(self.size)
        
    
    def Delete(self, value:int) -> None:
        i = self.heap.index(value)
        self.swap(self.heap, i,self.size)
        self.size -= 1
        self.heap.pop()
        self.Swap_down(i)
        
m = MINHEAP()
m.insert(20)
m.insert(54)
m.insert(32)
m.insert(21)
m.insert(76)
m.insert(12)
m.insert(127)
m.insert(162)
m.insert(42)
m.insert(70)
m.insert(15)
m.insert(78)
m.insert(65)
m.Delete(20)
m.Delete(12)
m.Delete(127)
print(m.get_list )   
```
